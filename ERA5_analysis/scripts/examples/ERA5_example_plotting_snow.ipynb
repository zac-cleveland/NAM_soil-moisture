# Set Path to ERA5 data and variables to choose

```python
era5_path = '/glade/campaign/collections/rda/data/ds633.0/'
```

```python
varlistnames = ['snow albedo','snow density','snow depth','temperature of snow']
varlist = ['asn', 'rsn', 'sd', 'tsn']
```

```python
ds_list = []

for var in varlist:
    print(var)
    
    files = glob.glob(era5_path + 'e5.oper.an.sfc/' + #'e5.oper.fc.sfc.accumu/' +
                      '200?03' + '/*'+var+'*.nc')
    files.sort
    
    # print(files)
    
    ds_temp = xr.open_mfdataset(files)
    
    ds_list.append(ds_temp)
    
ds = xr.merge(ds_list)
    
```

```python
# grab land/sea mask for data
lsm = xr.open_mfdataset('/glade/campaign/collections/rda/data/ds633.0/'+
                        'e5.oper.invariant/197901/'+
                        'e5.oper.invariant.128_172_lsm.ll025sc.1979010100_1979010100.nc')
```

# Plot 2-D map of first time slice 1st Var

```python
# snow albedo
#lsm_sw.LSM.isel(time=0).plot()
ds_sw.ASN.isel(time=0).plot()

# 
plt.title('Snow Albedo')
plt.show()
plt.close()
```

# Plot 2-D map of first time slice 2nd Var

```python
# snow density
ds_sw.RSN.isel(time=0).plot()

# 
plt.title('Snow Density')
plt.show()
plt.close()
```

```python
# snow depth
ds_sw.SD.isel(time=0).plot()

# 
plt.title('Snow Depth')
plt.show()
plt.close()
```

```python
# temperature of snow layer
ds_sw.TSN.isel(time=0).plot()

# 
plt.title('Temperature of Snow Layer')
plt.show()
plt.close()
```

# Plot just the days in March

```python
# get just the days of march to make the x-axis
march_days = ds_sw.where(ds_sw['time'].dt.year.isin([2000]), drop=True).time

for year in range(2000, 2009):
    
    ds_year = ds_sw.where(ds_sw['time'].dt.year.isin([year]), drop=True)
    plt.plot(march_days, ds_year.ASN.mean(['latitude', 'longitude']), label=year)

plt.title('Snow Albedo')
plt.legend()
plt.show()
plt.close()
```
