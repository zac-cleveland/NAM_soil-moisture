# Set Path to ERA5 data and variables to choose

```python
era5_path = '/glade/campaign/collections/rda/data/ds633.0/'
```

```python
varlistnames = ['CAPE','total cloud cover','large scale rain rate',
                'accumulated large scale precipitation']
varlistinst = ['cape', 'tcc']
varlistacc = ['lsrr', 'lsp']
```

```python
ds_list = []

for var in varlistinst:
    print(var)
    
    files = glob.glob(era5_path + 'e5.oper.an.sfc/' + #'e5.oper.fc.sfc.accumu/' +
                      '200?06' + '/*'+var+'*.nc')
    files.sort
    
    # print(files)
    
    ds_temp = xr.open_mfdataset(files)
    #print(ds_temp)
    
    ds_list.append(ds_temp)
    
dsinst = xr.merge(ds_list)


ds_list = []

for var in varlistacc:
    print(var)

    # Initialize the files list
    files = []

    # Try each directory path
    try:
        files += glob.glob(era5_path + 'e5.oper.fc.sfc.instan/' + '200?06' + '/*' + var + '*.nc')
    except Exception as e:
        print(f"Error in e5.oper.fc.sfc.instan: {e}")

    try:
        files += glob.glob(era5_path + 'e5.oper.fc.sfc.accumu/' + '200?06' + '/*' + var + '*.nc')
    except Exception as e:
        print(f"Error in e5.oper.fc.sfc.accumu: {e}")
        print("Re-evaluate your life")  

    # Sort the files list
    files.sort()

    # Open and merge the datasets
    ds_temp = xr.open_mfdataset(files)
    ds_list.append(ds_temp)

# Merge all datasets in the list
dsacc = xr.merge(ds_list)   
```

# Extract points for cities

```python
# Define lat/lon coordinates for cities
city_coordinates = {
    'Tucson': {'lat': 32.2226, 'lon': -110.9747},  # Tucson, AZ, US
    'Albuquerque': {'lat': 35.0844, 'lon': -106.6504},  # Albuquerque, NM, US
    'El Paso': {'lat': 31.7619, 'lon': -106.4850},  # El Paso, TX, US
    'Chihuahua': {'lat': 28.6328, 'lon': -106.0691},  # Chihuahua, Mexico
    'Empalme': {'lat': 27.9679, 'lon': -110.8239},  # Empalme, Sonora, Mexico
    'Mazatlan': {'lat': 23.2395, 'lon': -106.4063}  # Mazatlán, Sinaloa, Mexico
}

# Create empty dictionaries to store data arrays for each city
inst_data = {}
acc_data = {}

# Loop through each city and extract the closest grid point
for city, coords in city_coordinates.items():
    # Use the sel method with 'nearest' to find the closest grid point
    inst_data[city] = dsinst_sw.sel(latitude=coords['lat'], longitude=coords['lon'], method='nearest')
    acc_data[city] = dsacc_sw.sel(latitude=coords['lat'], longitude=coords['lon'], method='nearest')
```

# Plot just the days in June

```python
# get just the days of june to make the x-axis
june_days = dsinst_sw.where(dsinst_sw['time'].dt.year.isin([2000]), drop=True).time
unique_days = june_days.groupby('time.day').first()

for year in range(2000, 2009):
    
    ds_year = dsinst_sw.where(dsinst_sw['time'].dt.year.isin([year]), drop=True)
    # Resample data to daily mean
    ds_daily_mean = ds_year.resample(time='1D').mean()
    
    plt.plot(unique_days, ds_daily_mean.CAPE.mean(['latitude', 'longitude']), label=year)

plt.title('Mean CAPE - SW')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
plt.close()
```

# Now just for the cities

```python
# get just the days of june to make the x-axis
june_days = inst_data['Tucson'].where(inst_data['Tucson']['time'].dt.year.isin([2000]), drop=True).time
unique_days = june_days.groupby('time.day').first()

for city in city_coordinates:
    for year in range(2000, 2009):
        
        ds_year = inst_data[city].where(inst_data[city]['time'].dt.year.isin([year]), drop=True)
        # Resample data to daily mean
        ds_daily_mean = ds_year.resample(time='1D').max()
    
        # Plot the daily mean
        plt.plot(unique_days, ds_daily_mean.CAPE, label=year)
    
    plt.title(f'Max CAPE - {city}')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))    
    plt.show()
    plt.close()
```

```python
# get just the days of june to make the x-axis
june_days = dsacc_sw.where(dsacc_sw['forecast_initial_time'].dt.year.isin([2000]), drop=True).forecast_initial_time
unique_days = june_days.groupby('forecast_initial_time.day').first()

for year in range(2000, 2009):
    
    ds_year = dsacc_sw.mean(dim='forecast_hour').where(dsacc_sw['forecast_initial_time'].dt.year.isin([year]), drop=True)
    # Resample data to daily mean
    ds_daily_mean = ds_year.resample(forecast_initial_time='1D').mean()
    
    plt.plot(unique_days, 3600*12*ds_daily_mean.LSRR.mean(['latitude', 'longitude']), label=year)
    
   # plt.plot(june_days, 3600*12*ds_year.LSRR.mean(['latitude', 'longitude']), label=year)
   #     # *3600*12 to convert mm/s to mm/12hr

plt.title('large scale rain rate')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
plt.close()
```

```python
# get just the days of june to make the x-axis
june_days = acc_data['Tucson'].where(acc_data['Tucson']['forecast_initial_time'].dt.year.isin([2000]), drop=True).forecast_initial_time
unique_days = june_days.groupby('forecast_initial_time.day').first()

for city in city_coordinates:
    for year in range(2000, 2009):
        
        ds_year = acc_data[city].mean(dim='forecast_hour').where(acc_data[city]['forecast_initial_time'].dt.year.isin([year]), drop=True)
        # Resample data to daily mean
        ds_daily_mean = ds_year.resample(forecast_initial_time='1D').mean()
    
        # Plot the daily mean
        plt.plot(unique_days, 3600*24*ds_daily_mean.LSRR, label=year)
    
    plt.title(f'Large Scale Rain Rate - {city} \n averaged every 12 hr \n units: mm/day')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
    plt.close()
```

```python
# get just the days of june to make the x-axis
june_days = dsacc_sw.where(dsacc_sw['forecast_initial_time'].dt.year.isin([2000]), drop=True).forecast_initial_time
unique_days = june_days.groupby('forecast_initial_time.day').first()

for year in range(2000, 2009):
    
    ds_year = dsacc_sw.sum(dim='forecast_hour').where(dsacc_sw['forecast_initial_time'].dt.year.isin([year]), drop=True)
    # Resample data to daily mean
    ds_daily_mean = ds_year.resample(forecast_initial_time='1D').sum()
    
    plt.plot(unique_days, 1000*ds_daily_mean.LSP.mean(['latitude', 'longitude']), label=year)
        # *1000 to convert m to mm

plt.title('Accumulated Precipitation - SW \n mm/day')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
plt.close()
```
